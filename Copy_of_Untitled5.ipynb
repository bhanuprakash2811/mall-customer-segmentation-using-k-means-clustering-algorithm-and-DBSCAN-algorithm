```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("mall-customers-data.csv")

display(df.head())
display(df.corr(numeric_only=True))
```

```python
# distribution of annual income
sns.set(style="whitegrid")
sns.displot(df["annual_income"], kde=True, color="dodgerblue")

plt.title("Distribution of annual income (in k$)")
plt.xlabel("Range of annual income (in k$)")
plt.ylabel("Count")

plt.show()
```

```python
sns.set(style="whitegrid")
sns.displot(df["age"], kde=True, color="dodgerblue")

plt.title('Distribution of age', fontsize = 20)
plt.xlabel('Range of age')
plt.ylabel('Count')

plt.show()
```

```python
# distribution of spending score
sns.set(style="whitegrid")
sns.displot(df["spending_score"], kde=True, color="dodgerblue")

plt.title('Distribution of spending score', fontsize = 20)
plt.xlabel('Range of spending score (1-100)')
plt.ylabel('Count')

plt.show()
```

```python
# gender analysis
genders = df["gender"].value_counts()

colors = {
    "Male": "dodgerblue",
    "Female": "orange"
}

sns.set(style="whitegrid")
sns.barplot(x=genders.index, y=genders.values, palette=colors)

plt.show()
```

```python
# considering two features: annual income and spending score
X = df[["annual_income", "spending_score"]]
X.head()
```

```python
# scatterplot of the above data
sns.scatterplot(x="annual_income", y="spending_score", data=X, s=60, color="dodgerblue")

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Spending Score (1-100) vs Annual Income (k$)')

plt.show()
```

```python
wcss = []
for i in range(1, 11):
    k_means = KMeans(n_clusters=i, n_init="auto")
    k_means.fit(X)
    wcss.append(k_means.inertia_)

# the elbow curve
plt.plot(range(1, 11), wcss)
plt.plot(range(1, 11), wcss, linewidth=2.5, color="red", marker="8")

plt.title("Elbow Curve")
plt.xlabel("Number of clusters (values of k)")
plt.ylabel("WCSS values")
plt.xticks(np.arange(1, 11, 1))

plt.show()
```

```python
# taking k = 5
k_means_1 = KMeans(n_clusters=5, n_init="auto")

# fitting the input data
k_means_1.fit(X)

# predicting the labels of the input data
y_pred = k_means_1.predict(X)

# adding the labels to a column named label
df["label"] = y_pred

df.head()
```

```python
colors = ["red", "green", "dodgerblue", "gold", "purple"]

# scatterplot of the clusters
sns.scatterplot(
    x="annual_income",
    y="spending_score",
    hue="label",
    palette=colors,
    legend="full",
    data=df,
    s=60
)

plt.xlabel('Annual Income (in k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Spending Score (1-100) vs Annual Income (in k$)')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np
data= pd.read_csv("mall-customers-data.csv")
# 1. Prepare the data: Select features and scale them
features = data[['age', 'annual_income', 'spending_score']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 2. Apply DBSCAN clustering
# Setting eps and min_samples for a reasonable clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(scaled_features)

# Add cluster labels to the dataset
data['cluster'] = clusters

# 3. Display the number of points in each cluster and noise points
cluster_counts = data['cluster'].value_counts().sort_index()
print("Cluster counts (including noise):")
print(cluster_counts)

# 4. Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='annual_income', y='spending_score', hue='cluster', data=data, palette='viridis', style='gender')
plt.title("DBSCAN Clustering of Customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend(title="Cluster")
plt.show()
```
